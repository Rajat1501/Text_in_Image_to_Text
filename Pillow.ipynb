```python
from PIL import Image

img= Image.open("/Users/rajataggarwal/Desktop/simplee.png") #Loads image for PIL.Image
img.show() # opens the image in the default image viewer
img # displays the image in the code output itself
print(img) # prints details of the image
```

```python
img.filename #tells the filename of the image
img.format #tells the format of the image
img.mode #tells the mode of the image
img.size #tells the size of the image

```

```python
#Rotate and Resize the image
img.rotate(30).show() #rotates the image by 30 degrees
img.resize((400,250)).show() #resizes the image to 400x250
r1_image= img.rotate(90)
r1_image.show()
r1_image.size
```

```python
#Crop the image
img.crop((300,40,500,60)).show() # takes left, top, right, bottom
```

```python
#Adding text on a image
from PIL import ImageDraw
testing= Image.open("/Users/rajataggarwal/Desktop/simplee.png")
drawimagenew= ImageDraw.Draw(testing) #Loads image for PIL.ImageDraw draws on the image testing 
drawimagenew.text((50,90), "It is yoyoyo", (56,25,25)) # Where, what, color
testing
```

```python
#Padding an image 
width, height= img.size
print(width, height)
right= 20
left=20
top=20
bottom=50
new_width= width+ right + left
new_height= height+top+ bottom
paddedimage= Image.new(img.mode, (new_width, new_height), (0,0,255)) 
#Above, we created a new image with mode(color) of the original image i.e 'RGB' or if 'L' for grayscale,etc. Then dimensions of new image, and then color to fill image background.
#We have created a blue image of dimenions new_width and new_height
paddedimage.show() #shows padded image which is just blue as for now
paddedimage.paste(img, (30,40)) #this pastes img onto our created image
# this 30,40 ensures that original image is placed on new image such that it leaves space of left by 30, and top by 40 
paddedimage.show() 

```

```python
#Blurring an image
from PIL import ImageFilter
#We need to load image for ImageFilter. It works on Loaded image from Image.open itself i.e PIL.Image
#box blur= set each pixel as average value of all pixels in a box around it
imgboxblurred=img.filter(ImageFilter.BoxBlur(6)) #6 is the radius i.e no of pixels in one direction
imgboxblurred.show()    
#gaussian blur= Less weightage is given to pixels far away while averaging
imggaussianblurred=img.filter(ImageFilter.GaussianBlur(6))
imggaussianblurred.show()
#Simple BLUR filter
imgblurred=img.filter(ImageFilter.BLUR)
imgblurred.show()
```

```python
#Concatenating image
import matplotlib 
from matplotlib import pyplot as plt
# i will be concatenating 2 images,- img and imgblurred 
#1)resize them to equal size
a=img.resize((250,100))
b= imgblurred.resize((250,100))
newimage=Image.new('RGB', (500, 100),(1,1,1)) # creates a new image of size 500x100
newimage.paste(a, (0,0))
newimage.paste(b, (250,0))
#similarly you can concatenate images vertically, by changing the dimension.
img.show()
imgblurred.show()
newimage.show()

```
